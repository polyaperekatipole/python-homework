```python
from string import punctuation

def tokenize(s):
    words=s.lower().strip().split()
    twords = [word.strip(punctuation) for word in words]
    return twords
```

```python
import re

def transcription(s):
    pat='you'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'jʊ',s)
    pat='oo|ur'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ʊ',s)
    pat='sh|cio|tio'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ʃ',s)
    pat='ch'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'tʃ',s)
    pat='ank|ang'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'æŋ',s)
    pat='ore|or'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɔ',s)
    pat='ear'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɪ',s)
    pat='are|ar'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'a',s)
    pat='air'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɛ',s)
    pat='je|ge'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'d͡ʒi',s)
    pat='er$'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɹ',s)
    pat='ee|ea'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'i',s)
    pat='ay|ey'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'eɪ',s)
    pat='au|aw'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɔ',s)
    pat='oy|oi'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɔʊ',s)
    pat='ow|ou'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'aʊ',s)
    pat='y$'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ai',s)
    pat='ght'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'t',s)
    pat='th'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'θ',s)
    pat='ng'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ŋ',s)
    pat='ate'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'eɪ',s)
    pat='ace'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'eɪs',s)
    pat='a'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ə',s)
    pat='ice'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'aɪs',s)
    pat='ire'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'aɪ',s)
    pat='i'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɪ',s)
    pat='de'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'di',s)
    pat='die'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'daɪ',s)
    pat='r'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɹ',s)
    pat='ove'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'ɑv',s)
    pat='who'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'hʊ',s)
    pat='wh'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'w',s)
    pat='kn'
    findpat = len(re.findall(pat,s))
    if findpat==True:
        s = re.sub(pat,'n',s)
    return s
```

```python
alice=open('alice.txt').read()
a=alice.find('*** START OF THIS PROJECT GUTENBERG EBOOK') 
z=alice.find('*** END OF THIS PROJECT GUTENBERG EBOOK') 
alice_new=alice[a:z]
alice_tokens=tokenize(alice_new)
print(alice_tokens)
```

```python
translist=[]
for item in alice_tokens:
    transwords=transcription(item)
    translist.append(transwords)  
    
print(translist)
    
```

```python
dictionary = {} 
for key in alice_tokens: 
    for value in translist: 
        dictionary[key] = value 
        translist.remove(value) 
        break
```
